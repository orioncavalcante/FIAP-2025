```python
# Ex. 15

valor = float(input("Digite o valor da compra: "))
tipo_pagamento = int(input("Digite o número correspondente para seu tipo de pagamento:\r[1] Dinheiro ou Pix\r[2] Débito\r[3] Crédito"))

if tipo_pagamento == 1:
    valor_desconto1 = 10/100 * valor
    valor_final1 = valor - valor_desconto1
    print(f"Você recebeu 10% de desconto! Valor Inicial: R${valor}.\rValor com desconto: {valor_final1}")

elif tipo_pagamento == 2:
    valor_desconto2 = 5/100 * valor
    valor_final2 = valor - valor_desconto2
    print(f"Você recebeu 5% de desconto! Valor Inicial: R${valor}.\rValor com desconto: {valor_final2}")

elif tipo_pagamento == 3:
    valor_desconto3 = 5/100 * valor
    valor_final3 = valor + valor_desconto3
    print(f"Você pagará 5% a mais! Valor Inicial: R${valor}.\rValor total: {valor_final3}")
```

```python
# Ex. 16

usuario_correto = "admin"
senha_correta = "admin123**"


usuario = input("Digite seu usuário: ")
senha = input("Digite sua senha: ")

if usuario != usuario_correto and senha != senha_correta:
    print("Login inválido! Você tem mais duas tentativas!")

    usuario = input("Digite seu usuário: ")
    senha = input("Digite sua senha: ")

    if usuario != usuario_correto and senha != senha_correta:
        print("Login inválido! Você tem mais uma tentativas!")

        usuario = input("Digite seu usuário: ")
        senha = input("Digite sua senha: ")

        if usuario != usuario_correto and senha != senha_correta:
            print("Login inválido! Você não tem mais tentativas! Acesso bloqueado.")

        else:
            print("Login realizado com sucesso!")
    else:
        print("Login realizado com sucesso!")
else:
    print("Login realizado com sucesso!")
        

    
```
